```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = pd.read_csv('IRIS-Copy1.csv')
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(ax=axes[0, 0], data=iris, x='sepal_length', kde=True)
sns.histplot(ax=axes[0, 1], data=iris, x='sepal_width', kde=True)
sns.histplot(ax=axes[1, 0], data=iris, x='petal_length', kde=True)
sns.histplot(ax=axes[1, 1], data=iris, x='petal_width', kde=True)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.boxplot(ax=axes[0, 0], data=iris, x='sepal_length')
sns.boxplot(ax=axes[0, 1], data=iris, x='sepal_width')
sns.boxplot(ax=axes[1, 0], data=iris, x='petal_length')
sns.boxplot(ax=axes[1, 1], data=iris, x='petal_width')

```
