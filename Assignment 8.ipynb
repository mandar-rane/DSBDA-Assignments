```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')
```

```python
sns.catplot(x='sex', kind='count', hue='survived', data=titanic)
```

```python
sns.catplot(x='class', kind='count', hue='survived', data=titanic)
```

```python
sns.displot(data=titanic, x='age', hue='class', multiple='stack')
```

```python
sns.histplot(data=titanic, x='fare', bins=30)
```
